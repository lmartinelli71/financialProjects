```
#cargamos la librería camelot
import camelot
```

```
#se crea una tabla con la tabla de datos extraída del pdf
tables = camelot.read_pdf('/content/ipc_01_23891D383E4F.pdf', pages="12", flavor="stream", suppress_stdout=True)
```

```
#se observa la tabla extraída
df1=tables[0].df
df1
```

```
#se borran las filas 0 y 2 y las que tienen valores generales (que no utilizaremos)
df2=df1.drop([0,2,3,4,5,13,16,19,22,26,28,33,40,42,48,50], axis=0)
df2.reset_index(drop=True, inplace=True)
df2
```

```
#reemplazar nombres de columnas por la primer fila

df2=df2.rename(columns=df2.iloc[0]).drop(df2.index[0])
df2.reset_index(drop=True, inplace=True)
df2
```

```
#renombro la primer columna
df3=df2.rename(columns={'': 'Producto'})
df3
```

```
#se verifica individualmente, por ejemplo
df3.iloc[17,0]
```

```
#se coluca el nombre del producto en las columnas que tienen numeros y estan vacias
df3.iloc[18,0] = df3.iloc[17,0] + ' ' + df3.iloc[19,0]
df3.iloc[24,0] = df3.iloc[23,0] + ' ' + df3.iloc[25,0]
df3.iloc[30,0] = df3.iloc[29,0] + ' ' + df3.iloc[31,0]
df3
```

```
#se borran las filas que no contienen datos
df4=df3.drop([17,19,23,25,29,31], axis=0)
df4.reset_index(drop=True, inplace=True)
df4
```

```
#se verifica si hay datos faltantes
df4.isna().sum()
```

```
#se verifican los tipos de datos
df4.dtypes
```

```
import pandas as pd
```

Vamos a convertir los datos de tipo objeto a numericos

```
# Reemplazamos comas por puntos en todas las columnas
df4 = df4.replace(',', '.', regex=True)
```

```
# Convertimos las columnas de tipo object a tipo float excepto la columna Producto que no es numerica
for col in df4.columns:
    if df4[col].dtype == 'object' and col != 'Producto':
        df4[col] = pd.to_numeric(df4[col], errors='coerce')
```

2) **Cree una base de datos SQL y almacene el DataFrame final obtenido en el punto anterior como tabla**

```
#se crea la base de datos SQL
import sqlite3 as sql
```

```
#se crea el enlace a la base de datos
enlace = sql.connect("IPC.db")
```

```
import pandas.io.sql as pd_sql

#se almacena el DataFrame como tabla en la base creada
pd_sql.to_sql(df4,'IPC',enlace, index=False)

#se crea la base de datos IPC
# enlace es la conexión a la base IPC, index=False para no almacenar el índice de Python
```
